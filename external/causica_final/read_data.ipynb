```python
import json
import os
import numpy as np
```

```python
SRC_DIR = os.path.join("../../result_d4")
DST_DIR = os.path.join("data")

os.makedirs(DST_DIR, exist_ok=True)

for fname in os.listdir(SRC_DIR):
    if not fname.endswith(".json"):
        continue

    src_path = os.path.join(SRC_DIR, fname)

    with open(src_path, "r") as f:
        obj = json.load(f)

    X = np.asarray(obj["X_truth"], dtype=float)

    dataset_name = os.path.splitext(fname)[0]
    dataset_dir = os.path.join(DST_DIR, dataset_name)
    os.makedirs(dataset_dir, exist_ok=True)

    out_csv = os.path.join(dataset_dir, "all.csv")
    np.savetxt(out_csv, X, delimiter=",")

    print(f"[OK] {fname} -> {out_csv}, shape={X.shape}")
```
